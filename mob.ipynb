```python
df.drop(columns=['Device Model','User ID','Gender'],axis=1,inplace=True)
df.head(2)
```

```python
for graph in df.columns:
    plt.scatter(df[graph],df['Data Usage (MB/day)'])
    plt.xlabel(str(df[graph]))
    plt.ylabel(df['Data Usage (MB/day)'])
    plt.show()
```

```python
encoder = LabelEncoder()
df['Operating System']=encoder.fit_transform(df['Operating System'])
```

```python
from sklearn.model_selection import train_test_split
X = df.drop('Data Usage (MB/day)',axis=1)
y = df['Data Usage (MB/day)']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state=44)
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.metrics import  r2_score
```

```python
dt = DecisionTreeRegressor()

dt.fit(X_train, y_train)

dt_pred = dt.predict(X_test)

print("score is",r2_score(y_test, dt_pred))
```

```python
from sklearn.model_selection import GridSearchCV

# Define the range of max_depth values to test
param_grid = {'max_depth': [2, 3, 4, 5, 6, 7, 8, 9, 10]}

# Set up the Decision Tree and GridSearchCV
dtree = DecisionTreeRegressor()
grid_search = GridSearchCV(estimator=dtree, param_grid=param_grid, cv=5, scoring='r2')

# Fit the grid search
grid_search.fit(X_train, y_train)

# Find the best depth and R² score
best_depth = grid_search.best_params_['max_depth']
best_score = grid_search.best_score_

print(f"Best max_depth: {best_depth}, Best R²: {best_score}")
```
